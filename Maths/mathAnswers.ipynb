```python
import sympy

# Define the variables
m, c = sympy.symbols('m c')
total_students = 100
# Define the equations
# m/5 = c/8 --> m = 5c/8
# 100 = m + c  
eq1 = m/5 - c/8
eq2 = m + c - total_students

# Solve the system of equations
soln = sympy.solve((eq1, eq2), (m, c))

# Print the solution
print(soln)

if soln[m]/ total_students > 1/3:
    print("Yes, more than one-third of students are taking the math course.")
else:
    print("No, one-third or fewer students are taking the math course.")

```

```python
def filter_outcomes(outcomes, condition):
    filtered_outcomes = filter(condition, outcomes)
    return list(filtered_outcomes)
```

```python
dice_outcomes = [(i, j) for i in range(1, 5) for j in range(1, 5)]
print(dice_outcomes)
# 1 sum is 5
sum_greater_is_5 = lambda outcome: sum(outcome) == 5
filtered_outcomes = filter_outcomes(dice_outcomes, sum_greater_is_5)
print("len:",len(filtered_outcomes))
# 2 sum is less than 4
sum_greater_less_than_4 = lambda outcome: sum(outcome) <= 4
filtered_outcomes = filter_outcomes(dice_outcomes, sum_greater_less_than_4)
print("len:",len(filtered_outcomes))
# 3 sum is bigger than 4
sum_greater_more_than_4 = lambda outcome: sum(outcome) >= 4
filtered_outcomes = filter_outcomes(dice_outcomes, sum_greater_more_than_4)
print("len:",len(filtered_outcomes))
```

```python
def is_transitive(relation):
    for a, b in relation:
        for c, d in relation:
            if b == c:
                if (a, d) not in relation:
                    return False
    return True

# Example usage:
relation = [(1, 2), (2, 3), (1, 3)]
print(is_transitive(relation))  # Output: True

non_transitive_relation = [(1, 2), (2, 3), (3, 1), (1, 4)]
print(is_transitive(non_transitive_relation))  # Output: False

def is_simetrical(relation):
    for (a, b) in relation:
        if (b, a) not in relation:
            return False
    return True

# Example usage:
relation = [(1, 2), (2, 1), (1, 3), (3, 1)]
print(is_simetrical(relation))  # Output: True

non_transitive_relation = [(1, 2), (2, 3), (3, 1), (1, 4)]
print(is_simetrical(non_transitive_relation))  # Output: False
```

```python
import random

def random_partition(list, num_subsets):
    random.shuffle(list)
    partition = [set() for _ in range(num_subsets)]
    
    for i, a in enumerate(list):
        partition[i % num_subsets].add(a)
    
    return partition

data_list = [1, 2, 3, 4, 5, 6]
num_subsets = 4

partition = random_partition(data_list, num_subsets)

pairs_array = []

for i, subset in enumerate(partition):
    print(f"Subset {i + 1}: {subset}")
    pairs_array = pairs_array + get_pairs(subset)

print(pairs_array)

matrix = [  [1 if (a,b) in pairs_array else 0 for b in data_list] for a in data_list]

print(matrix)

print_matrix_with_headers(data_list, matrix, 5)

print(is_simetrical(pairs_array))
print(is_transitive(pairs_array))
```

```python
import math


def calculate_permutations(n, r):
    combinations = math.factorial(n) / ( math.factorial(n - r))
    return combinations

def calculate_combinations(n, r):
    combinations = math.factorial(n) / (math.factorial(r) * math.factorial(n - r))
    return combinations

def calculate_combination_with_repetition(n, r):
    combination = math.comb(n + r - 1, r)
    return combination

def calculate_total_combinations(n, as_array=False ):

    results = np.array([])
    for r in range(0, n+1):
        combinations = calculate_combinations(n, r)
        results = np.append(results, combinations)
        #total_combinations += combinations

    return  results.sum() if as_array == False  else results

# Example usage
result = calculate_total_combinations(3)
print("The total number of combinations is:", result)
```
