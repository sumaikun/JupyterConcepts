```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

# Set up the Venn diagram
fig, ax = plt.subplots()
venn = venn3(subsets = (19,32,26,35,10,8,10), ax=ax)
venn.get_label_by_id('A').set_text('French')
venn.get_label_by_id('B').set_text('Business')
venn.get_label_by_id('C').set_text('Music')
# Add circles
venn3_circles(subsets = (19,32,26,35,10,8,10), ax=ax)

# Add the number outside the circles
plt.text(x=0.9, y=0.5, s='51', fontsize=8)

# Add title
ax.set_title('Number of Students Taking French, Business, and Music Classes')

plt.show()
```

```python
import sympy

# Define the variables
m, c = sympy.symbols('m c')
total_students = 100
# Define the equations
eq1 = m/5 - c/8
eq2 = m + c - total_students

# Solve the system of equations
soln = sympy.solve((eq1, eq2), (m, c))

# Print the solution
print(soln)

if soln[m]/ total_students > 1/3:
    print("Yes, more than one-third of students are taking the math course.")
else:
    print("No, one-third or fewer students are taking the math course.")

```

```python
import itertools

A = {'a', 'b', 'c', 'd'}

# Generate all subsets of A
subsets = []
for i in range(len(A)+1):
    for combo in itertools.combinations(A, i):
        subsets.append(set(combo))

# Count the number of proper subsets
num_proper_subsets = len(subsets) - 1

print("All subsets of A:", subsets)
print("Number of proper subsets of A:", num_proper_subsets)
```

```python
#Using Set notation

# Set U
U = { 1,10,11,12 }

# Set X and Set Y
X = { 2,3,4,5,6,7 }
Y = { 2,9,0,8,7,6,5,4 }

# Print Set X
print("Set X:", X)

# Print Set Y
print("Set Y:", Y)

# Union of Sets X and Y
union = X.union(Y)
print("Union of Sets X and Y:", union)

# Intersection of Sets X and Y
intersection = X.intersection(Y)
print("Intersection of Sets X and Y:", intersection)

# Cartesian product of Sets X and Y
cartesian_product = list(itertools.product(X, Y))
print("Cartesian product of Sets X and Y:", cartesian_product)

# Complement of Set X with respect to Universal Set U
complement = U.difference(X)
print("Complement of Set X with respect to Universal Set U:", complement)
```

```python
# Using arrays 

U = [ 1,10,11,12 ]

# Set X and Set Y
X = [ 2,3,4,5,6,7,2,3,3,4 ]
Y = [ 2,9,0,8,7,6,5,4,4,4,2 ]

# Helper

def clean_list(list):
    clean = []
    for item in list:
        if item not in clean:
            clean.append(item)

    return clean

# Union

X = clean_list(X)
Y = clean_list(Y)

union = []

for num in X:
    if num not in union:
        union.append(num)

for num in Y:
    if num not in union:
        union.append(num)


print("union",union)

interception = []

for num in X:
    if num in Y and num not in interception:
        interception.append(num)


print("interception",interception)

cartesian = []

for num in X:
    for num2 in Y:
        cartesian.append([num,num2])

cartesian = clean_list(cartesian)

print("cartesian",cartesian)

complement = []

for num in U:
    if  num not in X:
        complement.append(num)

print("complement",complement)
```
